3\. Создайте граф карате-клуба, используя готовую функцию из `networkx`. Добавьте узлам атрибут `age`, заполненный целыми числами от 20 до 50. Визуализируйте граф, раскрасив узлы в цвет клуба (красный для "Mr. Hi" и зеленый для "Officer") и сделав размер узла пропорциональным величине атрибута `age`.

- [ ] Проверено на семинаре

```python
import networkx as nx
import matplotlib.pyplot as plt

# Создание графа карате-клуба
G = nx.karate_club_graph()

# Добавление атрибута "age" к каждому узлу
age_values = list(range(20, 51))
for node, age in zip(G.nodes, age_values):
    G.nodes[node]['age'] = age

# Создание списка цветов для узлов
node_colors = ['red' if G.nodes[node]['club'] == 'Mr. Hi' else 'green' for node in G.nodes]

# Создание списка размеров узлов на основе атрибута "age"
node_sizes = [G.nodes[node]['age'] for node in G.nodes]

# Визуализация графа
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=node_sizes, node_color=node_colors, edge_color='gray', alpha=0.7)
```

4\. Создайте ориентированный граф "путь" из 3 узлов. Визуализируйте граф со стрелками увеличенного размера. Визуализируйте граф, развернув стрелки в обратном направлении. На всех визуализациях добавляйте подписи узлов.

- [ ] Проверено на семинаре

```python
import networkx as nx
import matplotlib.pyplot as plt

# Создание ориентированного графа "путь" из 3 узлов
G = nx.DiGraph()
G.add_edges_from([(1, 2), (2, 3)])

# Визуализация графа с увеличенными стрелками
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, arrowsize=20)

# Добавление подписей узлов
labels = {1: 'A', 2: 'B', 3: 'C'}
nx.draw_networkx_labels(G, pos, labels)

# Отображение графа
plt.show()
```

```python
# Создание ориентированного графа "путь" из 3 узлов
G = nx.DiGraph()
G.add_edges_from([(1, 2), (2, 3)])
```

```python
# Визуализация графа с увеличенными стрелками
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, arrowsize=20)
```

```python
# Визуализация графа с развернутыми стрелками в обратном направлении
nx.draw_networkx(G, pos, with_labels=True, arrowsize=20, arrowstyle='<-, head_length=0.6, head_width=0.3')
```

5\. Создайте граф `Les Miserables`. Для каждого узла зафиксируйте координату на плоскости при помощи функции `random_layout`. Отрисуйте узлы графа при помощи функции `draw_networkx_nodes` с указанием размера узлов 10. Получите список `edge_colors` значений атрибута `weight` на ребрах и ограничьте (clip) его сверху значением 10, снизу значением 1. Отрисуйте ребра графа при помощи функции `draw_networkx_edges`, раскрасив ребра в цвет, соответствующий значениям `edge_colors` с применением цветовой карты `viridis`. Добавьте на визуализацию colorbar, показывающий градиент цветов ребер.

- [ ] Проверено на семинаре

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
# Создание графа Les Miserables
G = nx.les_miserables_graph()
```

```python
# Зафиксировать координаты узлов на плоскости с помощью функции random_layout
pos = nx.random_layout(G)
```

```python
# Получение списка значений атрибута weight на ребрах и ограничение сверху и снизу с помощью функции clip
edge_colors = [np.clip(G.edges[u, v]['weight'], 1, 10) for u, v in G.edges]
```

```python
# Отрисовка ребер графа с раскрашиванием в цвета, соответствующие значениям edge_colors
edges = nx.draw_networkx_edges(G, pos, edge_color=edge_colors, edge_cmap=plt.cm.viridis)
# Добавление colorbar
plt.colorbar(edges)
# Отображение графа
plt.show()
```
